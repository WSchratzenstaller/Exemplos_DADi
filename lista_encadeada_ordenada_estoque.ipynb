```python
# Exercício completo

from datetime import datetime

class No_lista:
  def __init__(self, nome, quantidade, validade):
    self.__nome = nome
    self.__quantidade = quantidade
    self.__validade = validade
    self.__proximo = None

  def set_proximo(self, proximo):
      self.__proximo = proximo

  def set_quantidade(self, quantidade):
      self.__quantidade = quantidade

  def get_nome(self):
      return self.__nome

  def get_quantidade(self):
      return self.__quantidade

  def get_validade(self):
      return self.__validade

  def get_proximo(self):
      return self.__proximo

class Minha_lista_simples_ordenada:
  def __init__(self):
    self.__inicio = None

  def inserir(self, nome, quantidade, validade = None):
    novo = No_lista(nome, quantidade, validade)
    if novo:
      # Primeiro nó
      if not self.__inicio:
        self.__inicio = novo
        return
      # Inserir alfabeticamente antes do primeiro nó atual
      if self.__inicio.get_nome() > nome:
        novo.set_proximo(self.__inicio)
        self.__inicio = novo
        return
      if validade:
        # Inserir por validade antes do primeiro nó atual (caso o primeiro seja o mesmo produto)
        if self.__inicio.get_nome() == nome and self.__inicio.get_validade() > validade:
          novo.set_proximo(self.__inicio)
          self.__inicio = novo
          return
      # Procurar posição
      auxiliar = self.__inicio
      while auxiliar.get_proximo():
        if auxiliar.get_proximo().get_nome() > nome:
          novo.set_proximo(auxiliar.get_proximo())
          auxiliar.set_proximo(novo)
          return
        elif auxiliar.get_proximo().get_nome() == nome and validade:
          while auxiliar.get_proximo().get_nome() == nome:
            if auxiliar.get_proximo().get_validade() > validade:
              novo.set_proximo(auxiliar.get_proximo())
              auxiliar.set_proximo(novo)
              return
            auxiliar = auxiliar.get_proximo()
            if not auxiliar.get_proximo():
              # O último elemento da lista
              novo.set_proximo(auxiliar.get_proximo())
              auxiliar.set_proximo(novo)
              return
          # O último elemento da sequência daquele produto
          novo.set_proximo(auxiliar.get_proximo())
          auxiliar.set_proximo(novo)
          return
        auxiliar = auxiliar.get_proximo()
      auxiliar.set_proximo(novo)

  # Verificar se o produto existe na lista
  def buscar(self, nome):
    if not self.__inicio:
      return 0
    auxiliar = self.__inicio
    while auxiliar.get_proximo():
      if auxiliar.get_nome() == nome:
        return auxiliar.get_quantidade()
      auxiliar = auxiliar.get_proximo()
    if auxiliar.get_nome() == nome:
        return auxiliar.get_quantidade()
    return 0

  # Imprimir todos os dados da lista
  def imprimir_todos(self):
    if not self.__inicio:
      print("Lista Vazia!")
      return
    saida = ""
    auxiliar = self.__inicio
    while auxiliar.get_proximo():
      saida += "Produto: " + auxiliar.get_nome() + " quantidade: " + str(auxiliar.get_quantidade()) 
      if auxiliar.get_validade():
        saida += " validade: " + str(auxiliar.get_validade().strftime('%d/%m/%Y')) 
      saida += "\n"
      auxiliar = auxiliar.get_proximo()
    saida += "Produto: " + auxiliar.get_nome() + " quantidade: " + str(auxiliar.get_quantidade()) 
    if auxiliar.get_validade():
      saida += " validade: " + str(auxiliar.get_validade().strftime('%d/%m/%Y')) 
    print(saida)

  # Agrupar os produtos e imprimir quantidade de cada grupo
  def imprimir_agrupado(self):
    if not self.__inicio:
      print("Não há elementos para agrupar.")
      return

    saida = ""
    quantidade = 0
    auxiliar = self.__inicio
    quantidade = auxiliar.get_quantidade()
        
    while auxiliar.get_proximo():
      if auxiliar.get_nome() == auxiliar.get_proximo().get_nome():
        quantidade += auxiliar.get_proximo().get_quantidade()
      else:
        saida += "Produto: " + auxiliar.get_nome() + " quantidade: " + str(quantidade) + "\n"
        quantidade = auxiliar.get_proximo().get_quantidade()
      auxiliar = auxiliar.get_proximo()

    saida += "Produto: " + auxiliar.get_nome() + " quantidade: " + str(quantidade) + "\n"
    print(saida)
    return

  # Alterar quantidade de produto cadastrado
  def alterar_quantidade(self, nome, quantidade):
    if not self.__inicio:
      print("Não há elementos para alterar quantidade mínima.")
      return

    auxiliar = self.__inicio

    while auxiliar:
      if auxiliar.get_nome() == nome:
        auxiliar.set_quantidade(quantidade)
        return
      auxiliar = auxiliar.get_proximo()  
    print("Produto não encontrado para alterar quantidade mínima.")
  
  def reduzir_estoque(self, nome, quantidade):
    if not self.__inicio:
      print("Não há produtos cadastrados para remover.")
      return

    quantidade_disponivel = self.quantidade_disponivel(nome)
    if quantidade_disponivel == 0:
      print("Produto não encontrado na lista")
      return
    elif quantidade_disponivel < quantidade:
      print("Estão disponíveis apenas ", str(quantidade_disponivel))
      return

    auxiliar = self.__inicio
    while auxiliar:
      if auxiliar.get_nome() == nome:
        saldo = auxiliar.get_quantidade() - quantidade
        if saldo > 0:
          auxiliar.set_quantidade(saldo)
          return
        else:
          self.remover(nome)
          if saldo < 0:
            self.reduzir_estoque(nome, -saldo)
            return
      auxiliar = auxiliar.get_proximo()
    return

  def remover(self, remover):
    auxiliar = self.__inicio
    if auxiliar.get_nome() == remover:
      self.__inicio = self.__inicio.get_proximo()
      return

    while auxiliar.get_proximo():
      if auxiliar.get_proximo().get_nome() == remover:
        auxiliar.set_proximo(auxiliar.get_proximo().get_proximo())
        return
      auxiliar = auxiliar.get_proximo()

  def quantidade_disponivel(self, nome):
    quantidade_disponivel = 0
    auxiliar = self.__inicio
    while auxiliar:
      if auxiliar.get_nome() == nome:
        quantidade_disponivel += auxiliar.get_quantidade()
      auxiliar = auxiliar.get_proximo()
    return quantidade_disponivel

  def conferir_minimo_produto(self, nome, quantidade_disponivel):
    quantidade_minima = self.buscar(nome)
    if quantidade_disponivel < quantidade_minima:
      print("\nNecessário comprar", nome, "\nMínimo esperado:", quantidade_minima, "\nDisponível após o consumo:", quantidade_disponivel)

# Classe principal com menu
class Produto():
  def main():
    lista_estoque = Minha_lista_simples_ordenada()
    lista_minimo = Minha_lista_simples_ordenada()
    while True:
      op = int(input("Opções:\n"
                    "1 - Cadastrar nova remessa de produto.\n" +
                    "2 - Imprimir lista com quantidade mínima cadastrada.\n" +
                    "3 - Imprimir lista com estoque cadastrado [todos os itens].\n" +
                    "4 - Imprimir lista com estoque cadastrado [agrupado].\n"
                    "5 - Editar quantidade mínima.\n" +
                    "6 - Consumir produto.\n" +
                    "7 - Sair.\n" +
                    "Informe opção: \n"))
      
      if op is 1:
        nome = input("Informe o nome do produto:\n")
        quantidade = int(input("Informe a quantidade.\n"))
        validade = input("Informe a validade (d/m/yyyy).\n")
        
        lista_estoque.inserir(nome, quantidade, datetime.strptime(validade, "%d/%m/%Y").date())
        if not lista_minimo.buscar(nome):
          quantidade = int(input("Informe a quantidade mínima de " + nome + " para manter no estoque:\n"))
          lista_minimo.inserir(nome, quantidade)
      elif op is 2:
        lista_minimo.imprimir_todos()
      elif op is 3:
        lista_estoque.imprimir_todos()
      elif op is 4:
        lista_estoque.imprimir_agrupado()
      elif op is 5:
        nome = input("Informe o nome do produto que deseja alterar:\n")
        quantidade = int(input("informe nova quantidade mínima para o produto:\n"))
        if quantidade == 0:
          lista_minimo.remover(nome)
        else:
          lista_minimo.alterar_quantidade(nome, quantidade)
      elif op is 6:
        nome = input("Informe o nome do produto que deseja consumir:\n")
        quantidade = int(input("informe quantidade que deseja consumir:\n"))
        lista_estoque.reduzir_estoque(nome, quantidade)
        quantidade_disponivel = lista_estoque.quantidade_disponivel(nome)
        lista_minimo.conferir_minimo_produto(nome, quantidade_disponivel)
      elif op is 7:
        break
      else:
        print("Opção inválida.")

if __name__ == '__main__':
  Produto.main()
```
